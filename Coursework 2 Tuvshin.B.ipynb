```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as sm
import matplotlib.pyplot as plt
dt = pd.read_excel('Course work 2.xlsx', sheet_name = 'Нийт')
data1 = dt.head(-1)
avg = dt['Нийт'].tail(-28)
data2 = pd.read_excel('Course work 2.xlsx', sheet_name = 'Оюутнуудын нэмэлт мэдээлэл')
barcol = ['#1aacbd','#00b7a8','#3abd80','#7dbf4e','#bdb817','#ffa600']
piecolor = ['#dbf9ff','#abf5ed','#95eec0','#a7e081','#cfc93e','#ffa600']
piecol = ['#003f5c','#444e86','#955196','#dd5182','#ff6e54','#ffa600']
wedge_props = {"linewidth": 2}
text_props = {"weight": "bold"}
explode = [0, 0.2, 0, 0, 0]
```

```python
#1
student = data2['Оюутны код']
code = student.values.tolist()
listid = []
for i in code:
    a = int(i[1:3])
    if a == 17:
        listid.append('5-р курс')
    if a == 16:
        listid.append('6-р курс')
    if a == 18:
        listid.append('4-р курс')
    elif a == 19:
        listid.append('3-р курс')
data2['Курс'] = listid
data2
```

```python
#2
x0 = pd.DataFrame(data1['Ирц-10'])
y0 = pd.DataFrame(data1['Шалгалт-30'])
mod0 = sm.ols('y0 ~ x0', data1).fit()
print(mod0.summary())
```

```python
#3
x1 = pd.DataFrame(data2['Судалсан хичээлийн тоо'])
y1 = pd.DataFrame(data2['Голч дүн'])
mod1 = sm.ols('y1 ~ x1', data2).fit()
print(mod1.summary())
```

```python
#4
x2 = pd.DataFrame(data2['ЭЕШ оноо'])
mod2 = sm.ols('y1 ~ x2', data2).fit()
print(mod2.summary())
print(mod2.params)
x3 = list(data2['ЭЕШ оноо'])
y2 = list(data2['Голч дүн'])
plt.scatter(x2, y1, color = '#003f5c')
plt.xlabel('ЭЕШ оноо', fontweight = 'bold')
plt.ylabel('Голч дүн', fontweight = 'bold')
z = np.polyfit(x3, y2, 1)
p = np.poly1d(z)
plt.plot(x2,p(x2),"--", color = '#d45087')
plt.show()
```

```python
#5
df = pd.DataFrame(mod2.params)
df1 = df.transpose()
reg = df1['Intercept'] + df1['x3'] * 620
reg
```

```python
#6
plt.figure(figsize = (8,4))
plt.bar('Оюутны нэр', 'Нийт', data = data1, color = barcol)
plt.xticks(rotation = 90)
#7
plt.hlines(avg, 0 ,28, color = 'purple', label = 'Дундаж')
plt.legend()
plt.title('Сурагчдын дүн', fontweight = 'bold')
plt.xlabel('Оюутны нэр', fontweight = 'bold')
plt.ylabel('Улирлын дүн', fontweight = 'bold')
```

```python
#8
ratings = data1.iloc[0:5, 12:14].set_axis(['Давтамж', 'Үнэлгээ'], axis = 1)
plt.figure(figsize = (20,5))
plt.pie('Давтамж', data = ratings, labels = 'Үнэлгээ', autopct = "%.f", colors = piecolor, shadow = True, wedgeprops = wedge_props, explode = explode)
plt.title('Үнэлгээний давтамж', fontweight = 'bold', bbox={'facecolor': 'grey', 'alpha': 0.2, 'pad': 8})
```

```python
#9
year = pd.DataFrame(data2['Суралцаж буй мэргэжил'].value_counts()).reset_index().rename(columns = {'index' : 'Суралцаж буй мэргэжил', 'Суралцаж буй мэргэжил': 'Давтамж'})
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (30, 6))
ax1.pie('Давтамж', labels = 'Суралцаж буй мэргэжил', data = year, autopct = "%.f", colors = piecolor, shadow = True)
ax1.set_title('Суралцаж буй мэргэжил', fontweight = 'bold', bbox={'facecolor': 'grey', 'alpha': 0.2, 'pad': 8})
ax2.bar('Суралцаж буй мэргэжил', 'Давтамж', data = year, color = barcol)
ax2.set_title('Суралцаж буй мэргэжил', fontweight = 'bold')
ax2.set_xlabel('Мэргэжил', fontweight = 'bold')
ax2.set_ylabel('Давтамж', fontweight = 'bold')

```
