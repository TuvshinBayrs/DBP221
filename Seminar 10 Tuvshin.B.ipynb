```python
import numpy as np 
x = np.arange(0, 10)
y = x*2
z = x**2
import matplotlib.pyplot as plt 
%matplotlib inline

#1 
plt.plot(x, y)
plt.title('Граф1')
plt.xlabel('x')
plt.ylabel('y')



```

```python
fig, ax = plt.subplots()
left, bottom, width, height = [.30, 0.6, 0.15, 0.15]
ax_new = fig.add_axes([left, bottom, width, height])
ax.plot(x, y, color = 'orange')
ax_new.plot(x, y)
```

```python
#2
x2 = np.linspace(0, 8, 100)
y2 = np.exp(x2)
fig, (ax1,ax2) = plt.subplots(1,2, figsize = (10, 2))
ax1.plot(x, y, '--b')
ax2.plot(x2, y2, color = "red")
ax2.set_xticks([0,2,4,6,8])
```

```python
x2 = np.linspace(0, 8, 100)
y2 = np.exp(x2)
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2, figsize = (10,4))
plt.subplots_adjust(wspace = 0.4 , hspace = 0.4)
ax1.plot(x, y, '--b')
ax2.plot(x2, y2, color = "red")
ax2.set_xticks([0,2,4,6,8])
ax3.plot(x, y, '--b')
ax4.plot(x2, y2, color = "red")
ax2.set_xticks([0,2,4,6,8])
```
