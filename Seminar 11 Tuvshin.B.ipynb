```python
import matplotlib.pyplot as plt 
%matplotlib inline
import pandas as pd
import numpy as np
#TIOBE Index
data = pd.read_excel('Seminar.xlsx', sheet_name = 0)
data1 = data.loc[::-1]
data1_des = data1.sort_values('Percentage', ascending = False)
#Adjustments
explode = (0,0,0,0,0,0.11)
colors = ['green','lightgreen', 'aqua','steelblue','lightsteelblue','lightblue']
piecol = ['#003f5c','#444e86','#955196','#dd5182','#ff6e54','#ffa600']
#Subplots
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (14,5))
#Bar chart
ax1.bar('Languages', 'Percentage', data = data1_des, color = colors)
ax1.set_title('TIOBE Index \n Dec 2021', fontsize = 12, fontweight = 'bold')
ax1.set_ylabel('Ratings (percent)', fontsize = 12, fontweight = 'bold')
ax1.set_xlabel('Languages', fontsize = 12, fontweight = 'bold')
#Pie chart
ax2.pie('Percentage', labels = 'Languages', 
        data = data1, explode = explode, 
        colors = piecol,autopct='%.1f%%',
       shadow = True, startangle = 100)
ax2.set_title('TIOBE Index, Dec 2021',
              bbox={'facecolor': 'grey', 'alpha': 0.2, 'pad': 8}, 
              fontsize = 12, fontweight = 'bold')
#PYPL Index
pypdata = pd.read_excel('Seminar.xlsx', sheet_name = 1)
mydata = pypdata.loc[::-1]
data_des = mydata.sort_values('Percentage', ascending = False)
#Adjustments
barcol = ['#1aacbd','#00b7a8','#3abd80','#7dbf4e','#bdb817','#ffa600']
piecolor = ['#dbf9ff','#abf5ed','#95eec0','#a7e081','#cfc93e','#ffa600']
#Subplots
fig, (ax3, ax4) = plt.subplots(1,2, figsize = (14,5))
ax3.bar('Languages', 'Percentage', data = data_des, color = barcol)
ax3.set_title('PYPL Index \n Dec 2021', fontsize = 12, fontweight = 'bold')
ax3.set_ylabel('Share (percent)', fontsize = 12, fontweight = 'bold')
ax3.set_xlabel('Languages', fontsize = 12, fontweight = 'bold')
#Pie chart
ax4.pie('Percentage', labels = 'Languages', 
        data = mydata, explode = explode, 
        colors = piecolor, autopct='%.1f%%',
       shadow = True, startangle = 100)
ax4.set_title('PYPL Index, Dec 2021',
              bbox={'facecolor': 'grey', 'alpha': 0.2, 'pad': 8}, 
              fontsize = 12, fontweight = 'bold')
```
